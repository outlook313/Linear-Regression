<b>Simple Linear Regression</b>

In simple linear regression, we aim to reveal the relationship between a single independent variable<br>
or you can say input, and a corresponding dependent variable or output. We can discuss this in a simple line as<br>

<font color=blue>y = w0 +w1x</font>

Here, Y speaks to the output or dependent variable, w0 and w1 are two obscure constants that speak to the intercept and <br>coefficient that is slope separately

We can also discuss this in the form of a graph and here is a sample simple linear regression model graph.<br>
Thus, in this whole blog, you will get to learn so many new things about simple linear regression in detail.<br>


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data=pd.read_csv('D:\mca_habeeb\Machine_Learning\income_happiness.csv')
print("---------------DATA SET----------------")
print(data)
print("Size of The Data Set:")
print(data.shape)
```

```python
income=pd.DataFrame(data['income'])
happiness=pd.DataFrame(data['happiness'])
model=LinearRegression().fit(income,happiness)
print(model.coef_)
print(model.intercept_)
print("Evaluvate the model :",model.score(income,happiness))
```

```python
x=([4.8976,5.897,6.3222])
x=pd.DataFrame(x)
y=model.predict(x)
y=pd.DataFrame(y)
df=pd.concat([x,y],axis=1,keys=['new_income','predicted_happiness'])
print(df)
data.plot(kind='scatter',x='income',y='happiness')
#ploting regression line
plt.plot(income,model.predict(income),color='red',linewidth=2)
#ploting predict values
plt.scatter(x,y,color='blue')
plt.show()
```
